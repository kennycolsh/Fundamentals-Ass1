```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from mpl_toolkits.mplot3d import Axes3D
sns.set(style="darkgrid")
```

```python
#load the dataset (Note size is the amount of people)
tips = sns.load_dataset("tips")
tips
```

```python
#We can see from this that the higher the bill the higher the tip
#the scatterplot is the default in relplot
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
```

```python
#When we use the Category plots we can see Thursday seems to 
#be more of a Female night out.
#And Saturday is the busiest day.

sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips);
```

```python
#Here we can see the when a bill goes above 20 it seems to increase
sns.relplot(x="total_bill", y="tip", size="size", sizes=(30, 100), data=tips);
```

```python
#Here we can see more Dinners and a higher amount of money is spent.
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=tips);
```

```python
plt.plot(df['total_bill'], df['tip'], 'r.')
plt.ylabel('tip')
plt.xlabel('total_bill')
plt.title('Total Bill Vs Tip')
plt.grid(True)
plt.rcParams['figure.figsize']=[15,5]
```

```python
plt.plot(df['total_bill'], df['day'], 'go',label='Total Bill')  # green dots
plt.plot(df['tip'], df['day'], 'b*',label='Tipsl')  # blue stars
plt.figure(figsize=(10,7))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

x =df['total_bill']
y =df['tip']
z =df.index

ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('total_bill')
ax.set_ylabel('tip')
ax.set_zlabel('Indexl')
```
